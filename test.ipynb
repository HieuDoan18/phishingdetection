```python
import pandas as pd
```

```python
data = pd.read_csv("phishing.csv")
data.dropna()
data
```

```python
def format_text(text):
    if isinstance(text, str):
        new_text = text.replace('\n', ' ')
        new_text = " ".join(new_text.split())
    else:
        new_text = ''
    return new_text

def format_link(link):
    if isinstance(link, str):
        new_link = link.replace('\n', '|')
    else:
        new_link = ''
    return new_link
```

```python
for i in range(data.shape[0]):
    data['text'].iloc[i] = format_text(data['text'].iloc[i])
    data['link'].iloc[i] = format_link(data['link'].iloc[i])
    data['link'].iloc[i]
    data['img'].iloc[i] = format_link(data['img'].iloc[i])
    data['img'].iloc[i]
```

```python
data.to_csv('phishing.csv', index=False)
# data.drop(data.columns[0], axis=1, inplace=True)
data
```
